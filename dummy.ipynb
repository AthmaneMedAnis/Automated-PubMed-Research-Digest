```python
def search_pubmed(query, max_result=5):
    Entrez.email = "mohamedanisathmane@gmail.com"

    print(f"Searching for '{query}'...")
    search_handle = Entrez.esearch(
        db="pubmed",
        term=query,
        retmax=max_result
    )
    search_results = Entrez.read(search_handle)
    search_handle.close()

    id_list = search_results["IdList"]

    print(f"Fetching details for {len(id_list)} papers...")
    fetch_handle = Entrez.efetch(
        db="pubmed",
        id=id_list,
        rettype="medline",
        retmode="text"
    )

    records = Medline.parse(fetch_handle)
    records = list(records)   # Convert iterator to list
    fetch_handle.close()

    # Convert to DataFrame
    df = pd.DataFrame(records)

    return df
results = search_pubmed("Covid19 ARNM")
```
